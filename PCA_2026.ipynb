```python
# ===============================
# PCA FROM SCRATCH (15+ FEATURES)
# ===============================
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Generate dataset (100 samples, 15 features)
np.random.seed(42)
X = np.random.rand(100, 15) * 10
X
```

```python
# Step 2: Mean centering
mean = np.mean(X, axis=0)
X_centered = X - mean
X_centered
```

```python

# Step 3: Covariance matrix
cov_matrix = np.cov(X_centered.T)
cov_matrix
```

```python
# Step 4: Eigen decomposition
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
eigenvalues, eigenvectors
```

```python
# Step 5: Sort eigenvalues and eigenvectors
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]
```

```python
 #Step 6: Explained variance ratio
explained_variance_ratio = eigenvalues_sorted / np.sum(eigenvalues_sorted)
explained_variance_ratio
```

```python
# Step 7: Select top 2 principal components
k = 2
principal_components = eigenvectors_sorted[:, :k]

principal_components
```

```python
# Step 8: Project data
X_pca = np.dot(X_centered, principal_components)
X_pca
```

```python
# ===============================
# VISUALIZATIONS
# ===============================

# 2D PCA Projection
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1])
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Projection (15 Features → 2D)")
plt.grid()
plt.show()
```

```python
# Scree Plot
plt.figure(figsize=(8, 5))
plt.plot(np.cumsum(explained_variance_ratio), marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.grid()
plt.show()
```

```python
print("Explained Variance Ratio (first 5 PCs):")
print(explained_variance_ratio[:5])
```

```python
###PCA on IRIS dataset

import pandas as pd
from sklearn.decomposition import PCA

#Get the IRIS dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
data = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])

#prepare the data
x = data.iloc[:,0:4]

#prepare the target
target = data.iloc[:,4]

#Applying it to PCA function
#mat_reduced = PCA(n_components=2)

pca=PCA(n_components=3)
principalComponents = pca.fit_transform(x)

#Creating a Pandas DataFrame of reduced Dataset
principal_df = pd.DataFrame(principalComponents , columns = ['PC1', 'PC2','PC3'])

#Concat it with target variable to create a complete Dataset
principal_df = pd.concat([principal_df , pd.DataFrame(target)] , axis = 1)
```

```python


import seaborn as sb
import matplotlib.pyplot as plt

plt.figure(figsize = (6,6))
sb.scatterplot(data = principal_df , x = 'PC1',y = 'PC2' , hue = 'target' , s = 60 , palette= 'icefire')
```
